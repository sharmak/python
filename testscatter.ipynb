```python
%matplotlib inline
import Quandl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from pylab import rcParams
from collections import OrderedDict
rcParams['figure.figsize'] = 15,20
plt.style.use('ggplot')
```

```python
ice_m3_df = Quandl.get("CHRIS/ICE_M3")
ice_m1_df = Quandl.get("CHRIS/ICE_M1")
```

```python
df = pd.concat([ice_m1_df['Settle'], ice_m3_df['Settle']], axis=1)
df.columns = ['ice_m3', 'ice_m1']
df = df.dropna()
df['year'] = [int(str(i)[:4]) for i in df.index.values]
df['month'] = [int(str(i)[5:7]) for i in df.index.values ]
df_end = df[-4:]
df_begin = df[:-4]
palette = sns.color_palette("dark", 12)
```

```python
markers = plt.Line2D.markers.keys()[::-1]
m_year  = dict()
c_month  = dict()
for i, year in enumerate(np.unique(df_begin.year)):
    df_year = df_begin[df_begin['year'] == year]
    m_year[year] = markers[i]
    for j, month in enumerate(np.unique(df_begin.month)):
        df_month = df_year[df_year['month'] == month]
        c_month[month] = palette[j]
        plt.scatter(df_month['ice_m3'], df_month['ice_m1'], marker=markers[i], color=palette[j], facecolor='none', s=20)
plt.plot(df.ice_m3.values, np.poly1d(np.polyfit(df.ice_m3.values, df.ice_m1.values, 1))(df.ice_m3.values))
size=120
for idx, row in df_end[::-1].iterrows():
    plt.scatter(row['ice_m3'], row['ice_m1'], marker=m_year[row['year']], color=c_month[row['month']],facecolor='none', s=size)
    size = size-20
plt.text(100, 5, '%s' %m_year[1997])
plt.show()
```
